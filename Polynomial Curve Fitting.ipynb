```python
#sample 10 points to be taken out of 100 from noisy sin(2*pi*x)
sample_points=10
x = np.linspace(0,1,num=sample_points)
eta=.4*np.random.normal(0,1,sample_points)
t=np.sin(2*np.pi*x)+eta

xf=np.linspace(0,1,num=100)
y=np.sin(2*np.pi*xf)

plt.plot(x,t,'bo', xf,y,'g-')
plt.legend(('Noisy Sampled Points','Sine Wave'))
```

```python
# Polynomial Curve Fitting
M=3
z=np.polyfit(x,t,M)
p=np.poly1d(z)
y=p(x)
plt.plot(x,t,'bo',x,y,'r-')
plt.legend(('Training Data','Model'))

err=0.5*(np.sum((y-t)**2))
print(err)
```

```python
M=9
z=np.polyfit(x,t,M)
p=np.poly1d(z)
y=p(x)
plt.plot(x,t,'bo',x,y,'r-')
plt.legend(('Training Data','Model'))

err=0.5*(np.sum((y-t)**2))
print(err)
```

```python
n=5
x_new=np.linspace(0,1,num=n)
eta=0.2*np.random.normal(0,1,n)
y_new=np.sin(2*np.pi*x_new)+eta

y_new_pred=p(x_new)

plt.figure
plt.plot(x,t,'bo',x,y,'r-',x_new,y_new_pred,'go')
plt.legend(('Traing Data','Model','Testing Data'))

err=0.5*(np.sum((y_new_pred-y_new)**2))
print(err)
```
