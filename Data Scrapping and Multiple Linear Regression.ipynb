```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url ='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
p = requests.get(url)
soup = BeautifulSoup(p.content, 'html.parser')
```

```python
p
```

```python
content = soup.find_all('div',class_ ='_13oc-S')
header = ["Product Name", "Ratings", "Price","Discount"]
itemfull = []
for item in content:
  items = []
  name = item.find('div', class_ = '_3Ztcd')
  ratings = item.find('div', class_ = '_3LWZlK')
  price = item.find('div', class_ = '_30jeq3 _1_WHN1')
  discount = item.find('div', class_ = '_3Ay6Sb')
  items.append(name.text)
  items.append(ratings.text)
  items.append(price.text)
  if(discount is not None):
    items.append(discount.text)
  else:
    items.append("No Discount")
  itemfull.append(items)

  print("Product Name:", name.text)
  print("Ratings:", ratings.text)
  print("Price:", price.text)
  if(discount is not None):
    print("Discount:", discount.text)

  with open('DataScrapMobiles.csv', 'w') as csvfile:
   csvwriter = csv.writer(csvfile)
   csvwriter.writerow(header)
   csvwriter.writerows(itemfull)
```

```python
import numpy as nm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/DataScrapMobiles.csv')
df
```

```python
encoding = 'utf-8',line_terminator='^',sep='~',index=False
```

```python
df['Price']=df['Price'].str.replace(',', '')
```

```python
df['Price'] = df['Price'].str.replace('%','')
```

```python
df['Price']
```

```python
X=df.drop('Price',axis=1)
y=['Price']
```

```python
X
```

```python
X=X.drop('Product Name',axis=1)
discount=pd.get_dummies(X,drop_first=True)
X=X.drop('Discount',axis=1)
X=pd.concat([X,discount],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df.iloc[:,2].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
ln.fit(X_train,y_train)
```

```python
y_pred = ln.predict(X_test)
```

```python
train_score = ln.score(X_train,y_train)
test_score=ln.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
score=r2_score(y_test,y_pred)
print(score)
MSE=mean_squared_error(y_test,y_pred)
print(MSE)
RMSE=nm.sqrt(MSE)
print(RMSE)
```
