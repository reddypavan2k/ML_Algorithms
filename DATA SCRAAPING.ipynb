```python
# scrap flipkart site mobile content into csv file
from bs4 import BeautifulSoup
import requests
import csv

url ='https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_3_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_3_0_na_na_na&as-pos=3&as-type=TRENDING&suggestionId=mobiles&requestId=7ced07a2-9830-4395-b4d0-a1918d8abafe'
p = requests.get(url)
soup = BeautifulSoup(p.content, 'html.parser')

content = soup.find_all('div',class_ ='_13oc-S')
header = ["Product Name", "Ratings", "Price","Discount"]
itemfull = []
for item in content:
  items = []
  name = item.find('div', class_ = '_4rR01T')
  ratings = item.find('div', class_ = '_3LWZlK')
  price = item.find('div', class_ = '_30jeq3 _1_WHN1')
  discount = item.find('div', class_ = '_3Ay6Sb')
  items.append(name.text)
  items.append(ratings.text)
  items.append(price.text)
  if(discount is not None):
    items.append(discount.text)
  else:
    items.append("No Discount")
  itemfull.append(items)

  print("Product Name:", name.text)
  print("Ratings:", ratings.text)
  print("Price:", price.text)
  if(discount is not None):
    print("Discount:", discount.text)

  with open('DataScrapMobiles.csv', 'w') as csvfile:
   csvwriter = csv.writer(csvfile)
   csvwriter.writerow(header)
   csvwriter.writerows(itemfull)
```
